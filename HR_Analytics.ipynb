```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set_style("whitegrid")
```

```python
overall_attrition=df['Attrition'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(overall_attrition, labels=['Stayed','Left'],autopct='%1.1f%%',colors=['#2ecc71','#e74c3c'])
plt.title('Overall Employee Attrition Rate')
plt.savefig('chart1_attrition_pie.png',bbox_inches='tight',dpi=150)
plt.show()
```

# Overall Attrition: 16% of employees left the company. This is higher than the industry average of 10 to 15%. It shows there is a retention issue that requires immediate attention.

```python
dept_attrition = df.groupby('Department')['Attrition'].value_counts(normalize=True).mul(100).rename('Percentage').reset_index()
plt.figure(figsize=(8,5))
sns.barplot(data=dept_attrition,x='Department',y='Percentage', hue='Attrition', palette=['#2ecc71','#e74c3c'])
plt.title('Attrition Rate by Department')
plt.ylabel('Percentage (%)')
plt.xlabel('Department')
plt.savefig('chart2_department_attrition.png', bbox_inches='tight', dpi=150)
plt.show()
```

# People are leaving the sales dept. and making problems. Sales department staff is leaving in droves. The number of people leaving the sales department is 21 percent. That's a lot because that means one in five people who work in the sales department are going to leave the sales department. The sales department is losing people than the HR department with a rate of 19% than the R&D department with a rate of 14% The sales department has the number of people leaving the sales .

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Attrition', y='MonthlyIncome',palette=['#e74c3c','#2ecc71'])
plt.title('Monthly Income vs Attrition')
plt.grid()
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.savefig('chart3_income_attrition.png', bbox_inches='tight', dpi=150)
plt.show()
```

# Salary is the biggest driver Employees who quit made on average 40% less than employees who stayed. The median income of leavers is almost 40% lower than that of retained employees. Compensation is the single biggest driver of attrition.

```python
plt.figure(figsize=(8,5))
sns.histplot(data=df, x='Age',hue='Attrition',multiple='stack',palette=['#e74c3c','#2ecc71'])
plt.title('Age Distribution vs Attrition')
plt.xlabel('Age')
plt.grid()
plt.ylabel('Number of Employees')
plt.savefig('chart4_age_attrition.png', bbox_inches='tight', dpi=150)
plt.show()
```

# The greatest amount of turnover is among those employees in the 20 to 30 age range.Young workers are quitting early, probably due to better opportunities elsewhere or unmet expectations for career growth.

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df, x='Attrition', y='YearsAtCompany',hue='Attrition',palette=['#e74c3c','#2ecc71'], legend=False)
plt.title('Years at Company vs Attrition')
plt.xlabel('Attrition')
plt.ylabel('Years at Company')
plt.savefig('chart5_years_attrition.png', bbox_inches='tight', dpi=150)
plt.show()
```

# Attrition happens early. Those who leave stay an average of 3 years, while those who stay average 7 years. It’s not a long term employee problem. It’s a new employee retention problem the company has.

```python
plt.figure(figsize=(8,5))
sns.countplot(data=df,x='JobSatisfaction',hue='Attrition', palette=['#2ecc71','#e74c3c'])
plt.title('Job Satisfaction vs Attrition')
plt.grid()
plt.xlabel('Job Satisfaction (1=Low, 4=High)')
plt.ylabel('Number of Employees')
plt.savefig('chart6_satisfaction_attrition.png', bbox_inches='tight', dpi=150)
plt.show()
```

# Impact on Job Satisfaction Employees with the lowest job satisfaction leave at 2X the rate (23%) of highly satisfied employees (11%).Salary and tenure, however, remain the stronger predictors of attrition overall.
